### Анализ пропусков и дубликатов в данных

```python
# Функция для анализа пропусков и дубликатов в данных
def missing_duplicates_report(df, name):
    print(f"\n{'='*50}")
    print(f"Выборка: {name}")
    print(f"{'='*50}")
    print(f"Размер: {df.shape[0]} строк, {df.shape[1]} столбцов")
    
    # Пропуски
    missing = df.isnull().sum()
    missing_pct = 100 * missing / len(df)
    missing_report = pd.DataFrame({'пропусков': missing, 'доля (%)': missing_pct})
    missing_report = missing_report[missing_report['пропусков'] > 0]
    if len(missing_report) > 0:
        print("Пропуски:")
        print(missing_report)
    else:
        print("Пропуски отсутствуют.")
    
    # Явные дубликаты (полностью одинаковые строки)
    explicit_dups = df.duplicated().sum()
    print(f"\nЯвные дубликаты (полностью совпадающие строки): {explicit_dups}")

    # Визуализация распределения пропусков
    sns.heatmap(df.isnull(), cbar=False, cmap='plasma')
    plt.title('Распределение пропусков в выборке')
    plt.show()
```

<b>Комментарий ревьюера v1:</b>

```python
# Вызываем функцию для обучающей и тестовой выборки
for data, name in [(train_data, 'Train'), (test_data, 'Test')]:
    missing_duplicates_report(data, name)
```

```python
# Функция для добавления метрик в список
def add_metrics(new_metric, name):
    # Проверяем наличие модели в списке
    model_idx = None
    for i, dct in enumerate(metrics_list):
        if dct['Model'] == name:
            model_idx = i
            break
    
    if model_idx is None:
        metrics_list.append(new_metric)
    else:
        metrics_list[model_idx] = new_metric
        
    metrics_list.sort(key=lambda x: x['RMSE'])


# Функция для кросс-валидации и рассчета метрик
def cross_valid_model(X, y, pipline, name_model, scoring, cv):
    # Кросс-валидация для модели
    model_scores = cross_validate(
        pipline, 
        X, 
        y, 
        cv=cv, 
        scoring=scoring,
        return_train_score=False)

    # Подготавливаем новые значения
    new_data = {
        'Model': name_model,
        'RMSE': round(np.sqrt(-model_scores['test_neg_mse'].mean()), 2),
        'MAE': round(-model_scores['test_neg_mae'].mean(), 2),
        'R2': round(model_scores['test_r2'].mean(), 2)
    }

    # Добавляем метрики в список
    add_metrics(new_data, name_model)
    
    return model_scores

# Отоброжение метрик
def show_metrics(name_model='all'):
    metrics_df = pd.DataFrame(metrics_list).sort_values(by='RMSE')
    if name_model=='all':
        display(metrics_df)
    else:
        display(metrics_df[metrics_df['Model']==name_model])

# Функция для визуализации предсказаний модели
def visual_predict(y_test, y_pred, name_model):
    plt.figure(figsize=(12, 6))
    plt.plot(y_test.values, label='Фактические значения')
    plt.plot(y_pred, label='Прогнозы модели', alpha=0.7)
    plt.title(f'Сравнение фактических значений и прогнозов для модели {name_model}')
    plt.xlabel('Временные шаги')
    plt.ylabel('Количество арендованных велосипедов')
    plt.legend()
    plt.show() 
```

### [Предобработка данных <a id=3.2></a>](#content)

```python
# Создаем кастомного трансформера для обработки признаков
class FeatureCreator(BaseEstimator, TransformerMixin):
    def __init__(self, 
                 temp_col='Temperature', 
                 solar_col='Solar Radiation (MJ/m2)', 
                 hum_col='Humidity(%)', 
                 wind_col='Wind speed (m/s)', 
                 vis_col='Visibility (10m)',
                 time_nithg='Time_Period_Night',
                 time_morn='Time_Period_Morning',
                 time_even='Time_Period_Evening',
                 time_late_even='Time_Period_Late Evening'):
        self.feature_names_out_ = None
        self.temp_col = temp_col
        self.solar_col = solar_col
        self.hum_col = hum_col
        self.wind_col = wind_col
        self.vis_col = vis_col
        self.time_nithg = time_nithg
        self.time_morn = time_morn
        self.time_even = time_even
        self.time_late_even = time_late_even
     
    def fit(self, X, y=None):
        return self
    
    def transform(self, X):        
        X_copy = X.copy()

        # Создаем новые признаке на основе числовых признаков
        X_copy['temp_humidity'] = X_copy[self.temp_col] * X_copy[self.solar_col]
        X_copy['temp_squared'] = X_copy[self.temp_col] **2
        X_copy['temp_solar'] = X_copy[self.temp_col] * X_copy[self.hum_col]
        X_copy['temp_wind'] = X_copy[self.temp_col] * X_copy[self.wind_col]

        # Создаем новые бинарный признаки
        X_copy['extreme_hot'] = (X_copy[self.temp_col] > 30).astype(int)
        X_copy['low_visibility'] = (X_copy[self.vis_col] < 1000).astype(int)

        # Объединение биннарных признаков в один категориальный признак
        X_copy['time_period'] = 'daytime'
        X_copy.loc[X_copy[self.time_nithg]==True, 'time_period'] = 'night'
        X_copy.loc[X_copy[self.time_morn]==True, 'time_period'] = 'morning'
        X_copy.loc[X_copy[self.time_even]==True, 'time_period'] = 'evening'
        X_copy.loc[X_copy[self.time_late_even]==True, 'time_period'] = 'late_evening'

        self.feature_names_out_ = X_copy.columns.tolist()

        return X_copy

    def get_feature_names_out(self):
        return self.feature_names_out_
```

<div class="alert alert-warning">
<b>Комментарий ревьюера v1:</b>

В трансформере ты жестко прописал старые названия столбцов (например, Solar Radiation (MJ/m2)). Это сработало, потому что ранее переименование не применилось к train_df. Но в будущем старайся передавать имена нужных колонок через параметры инициализации класса (__init__), чтобы трансформер был универсальным.

<div class="alert alert-success">
<b>Комментарий ревьюера v1:</b>

Написание собственного трансформера с наследованием от BaseEstimator и TransformerMixin — это показатель отличного уровня подготовки. Трансформер легко встроится в пайплайн.

```python
# Создание числового пайплайна
numeric_pipe = Pipeline(steps=[
    ('imputer', SimpleImputer(strategy='median')),
    ('scaler', StandardScaler())
])

# Создание категориального пайплайна
category_pipe = Pipeline(steps=[
    ('imputer', SimpleImputer(strategy='most_frequent')),
    ('one', OneHotEncoder(handle_unknown='ignore', sparse_output=False))
])

# Создаём ColumnTransformer с обработкой разных типов признаков
preprocessor = ColumnTransformer([
    ('num', numeric_pipe, make_column_selector(dtype_include=['int64', 'float64'])), 
    ('cat', category_pipe, make_column_selector(dtype_include=['object', 'category', 'string']))
], remainder='passthrough'
)

# Создание базового пайплайна
def create_pipeline(model):
    return Pipeline([
        ('feature_create', FeatureCreator()),
        ('preprocessor', preprocessor),
        ('model', model)
    ])
```

## [Оптимизация параметров <a id=4></a>](#content)

```python
# Задаем функцию-цель для оптимизации optuna
def objective(trial):
    # Выбираем модель для оптимизации
    model_name = trial.suggest_categorical('model', ['KNN_optuna', 'DecisionTree_optuna'])

    if model_name == 'KNN_optuna':
        params = {
            'n_neighbors': trial.suggest_int('n_neighbors', 1, 50),
            'weights': trial.suggest_categorical('weights', ['uniform', 'distance']),
            'metric': trial.suggest_categorical('metric', ['minkowski']),
            'p': trial.suggest_int('p', 1, 3)
        }

        pipeline = create_pipeline(
            KNeighborsRegressor(
                **params
            )
        )
    else:
        params = {
            "max_depth": trial.suggest_int('max_depth', 3, 30),
            "min_samples_split": trial.suggest_int('min_samples_split', 2, 50),
            "min_samples_leaf": trial.suggest_int('min_samples_leaf', 1, 20),
            "criterion": trial.suggest_categorical('criterion', ['squared_error', 'friedman_mse', 'absolute_error'])
        }

        pipeline = create_pipeline(
            DecisionTreeRegressor(
                **params,
                random_state=RANDOM_STATE
            )
        )
    
    kf = KFold(n_splits=5, shuffle=True, random_state=RANDOM_STATE)
    scores = cross_val_score(
        pipeline,
        X_train, 
        y_train,
        cv=kf,
        scoring='neg_mean_squared_error'
    )

    return -scores.mean()
```

<b>Комментарий ревьюера v1:</b>

В функции для Optuna есть критическая ошибка логики. Это условие всегда ложно, так как 'KNN_optuna' не равно 'KNN'. Из-за этого код всегда переходит в блок else и перебирает параметры для решающего дерева (посмотри на вывод Optuna: при выборе KNN_optuna подбираются max_depth и min_samples_split). Чтобы исправить, замени условие на if model_name == 'KNN_optuna':.

```python
# Запускем оптимизацию
study = optuna.create_study(
    direction='minimize',
    sampler=optuna.samplers.TPESampler(seed=3)
)
study.optimize(objective, n_trials=15, show_progress_bar=True)
```

```python
# Лучшие гиперпараметры и качество
best_params = study.best_params
best_value = study.best_value
best_model_name = best_params.pop('model')

print(f'Лучшая модель {best_model_name}')
print('Лучшие гиперпараметры:')
for key, value in best_params.items():
    print(f'     {key}: {value}')
print('Лучшее значение RMSE:', round(np.sqrt(best_value), 2))
```
